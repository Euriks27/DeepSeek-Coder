## 🧱 Célula 3: Clientes de API (BCB, FRED, World Bank)

```python
import pandas as pd
import numpy as np
import requests
import logging
import time
from datetime import datetime, timedelta
from functools import wraps

# ========== Decorator de Retry ==========
def retry(max_attempts=3, initial_delay=1, backoff_factor=2):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            delay = initial_delay
            for attempt in range(max_attempts):
                try:
                    return func(*args, **kwargs)
                except Exception as e:
                    if attempt == max_attempts - 1:
                        raise
                    time.sleep(delay)
                    delay *= backoff_factor
            return None
        return wrapper
    return decorator

# ========== BCB Client ==========
class BCBClient:
    BASE_URL = "https://api.bcb.gov.br/dados/serie/bcdata.sgs.{}/dados?formato=json"
    MAX_YEARS = 10

    def __init__(self):
        self.cache = {}

    @retry(max_attempts=5, backoff_factor=2)
    def _fetch_chunk(self, series_code, start_str, end_str):
        url = self.BASE_URL.format(series_code)
        params = []
        if start_str: params.append(f"dataInicial={start_str}")
        if end_str: params.append(f"dataFinal={end_str}")
        if params: url += "&" + "&".join(params)
        resp = requests.get(url, timeout=30)
        resp.raise_for_status()
        return resp.json()

    def _process(self, data):
        if not data: return pd.DataFrame()
        df = pd.DataFrame(data)
        df['data'] = pd.to_datetime(df['data'], format='%d/%m/%Y')
        df['valor'] = pd.to_numeric(df['valor'], errors='coerce')
        df = df.set_index('data').sort_index()
        return df

    def get_series(self, code, start_date=None, end_date=None):
        start_dt = pd.to_datetime(start_date) if start_date else None
        end_dt = pd.to_datetime(end_date) if end_date else datetime.now()
        all_data = []

        if start_dt and (end_dt - start_dt).days / 365.25 > self.MAX_YEARS:
            current = start_dt
            while current <= end_dt:
                chunk_end = min(current + pd.DateOffset(years=self.MAX_YEARS), end_dt)
                chunk_data = self._fetch_chunk(code,
                                               current.strftime('%d/%m/%Y'),
                                               chunk_end.strftime('%d/%m/%Y'))
                all_data.extend(chunk_data)
                current = chunk_end + pd.DateOffset(days=1)
        else:
            all_data = self._fetch_chunk(code,
                                         start_dt.strftime('%d/%m/%Y') if start_dt else None,
                                         end_dt.strftime('%d/%m/%Y') if end_dt else None)

        df = self._process(all_data)
        df.columns = [f'bcb_{code}']
        return df

    # Métodos de conveniência
    def get_selic(self, start=None, end=None):
        return self.get_series('4390', start, end)   # SELIC meta
    def get_ipca(self, start=None, end=None):
        return self.get_series('433', start, end)    # IPCA mensal
    def get_cambio(self, start=None, end=None):
        return self.get_series('1', start, end)      # BRL/USD
    def get_fsi(self, start=None, end=None):
        return self.get_series('24375', start, end)  # Fiscal Stress Index

# ========== FRED Client ==========
class FREDClient:
    BASE_URL = "https://api.stlouisfed.org/fred/series/observations"

    def __init__(self, api_key):
        self.api_key = api_key

    @retry(max_attempts=5, backoff_factor=2)
    def get_series(self, series_id, start=None, end=None, freq=None):
        params = {'series_id': series_id, 'api_key': self.api_key, 'file_type': 'json'}
        if start: params['observation_start'] = start
        if end: params['observation_end'] = end
        if freq: params['frequency'] = freq

        resp = requests.get(self.BASE_URL, params=params, timeout=30)
        resp.raise_for_status()
        data = resp.json()['observations']
        df = pd.DataFrame(data)
        df['date'] = pd.to_datetime(df['date'])
        df['value'] = pd.to_numeric(df['value'], errors='coerce')
        df = df.set_index('date').sort_index()
        df.columns = [series_id]
        return df

# ========== World Bank Client ==========
class WorldBankClient:
    def __init__(self):
        import wbgapi as wb
        self.wb = wb

    @retry(max_attempts=3)
    def get_indicator(self, indicator, country='BR', start=None, end=None):
        time_range = range(start, end+1) if start and end else 'all'
        df = self.wb.data.DataFrame(indicator, country, time=time_range)
        if df.empty: return pd.DataFrame()
        df = df.stack().reset_index()
        df.columns = ['series', 'year', 'value']
        df['date'] = pd.to_datetime(df['year'].astype(str))
        df = df.pivot(index='date', columns='series', values='value')
        return df

print("✅ Clientes de API definidos.")
```

## 💾 Célula 4: Banco de Dados SQLite (Simulando TimescaleDB)

```python
from sqlalchemy import create_engine, Column, Float, Date, String
from sqlalchemy.orm import sessionmaker, declarative_base
from sqlalchemy.dialects.sqlite import insert as sqlite_insert

# Usamos SQLite para simplicidade no Colab
engine = create_engine('sqlite:///macrolab.db')
Base = declarative_base()

class BCBData(Base):
    __tablename__ = 'bcb_indicators'
    date = Column(Date, primary_key=True)
    series_code = Column(String, primary_key=True)
    value = Column(Float)

Base.metadata.create_all(engine)
Session = sessionmaker(bind=engine)

def load_bcb_data_fast(df_transformed):
    """Upsert otimizado para SQLite."""
    if df_transformed.empty: return
    session = Session()
    for _, row in df_transformed.iterrows():
        stmt = sqlite_insert(BCBData).values(
            date=row['date'], series_code=row['series_code'], value=row['value']
        ).on_conflict_do_update(
            index_elements=['date', 'series_code'],
            set_=dict(value=row['value'])
        )
        session.execute(stmt)
    session.commit()
    session.close()

def transform_bcb_data(df, series_code):
    if df.empty: return pd.DataFrame()
    df = df.reset_index()
    df = df.rename(columns={'data': 'date', f'bcb_{series_code}': 'value'})
    df['series_code'] = series_code
    df['date'] = pd.to_datetime(df['date']).dt.date
    return df[['date', 'series_code', 'value']]

print("✅ Banco de dados SQLite configurado.")
```

```python
print("🌎 Coletando dados reais...")

# Inicializa clientes
bcb = BCBClient()
fred = FREDClient(api_key=FRED_API_KEY) if FRED_API_KEY else None
wb = WorldBankClient()

# Período de 10 anos
end_date = datetime.now().strftime('%Y-%m-%d')
start_date = (datetime.now() - timedelta(days=10*365)).strftime('%Y-%m-%d')

# BCB
df_selic = bcb.get_selic(start_date, end_date)
df_ipca = bcb.get_ipca(start_date, end_date)
df_cambio = bcb.get_cambio(start_date, end_date)
df_fsi = bcb.get_fsi(start_date, end_date)

# FRED (com fallback sintético)
if fred:
    df_fedfunds = fred.get_series('DFF', start_date, end_date, freq='m')  # Federal Funds Rate
    df_vix = fred.get_series('VIXCLS', start_date, end_date, freq='m')     # VIX
else:
    print("⚠️ FRED indisponível, gerando dados sintéticos.")
    idx = pd.date_range(start_date, end_date, freq='MS')
    df_fedfunds = pd.DataFrame({'DFF': 2.0 + np.random.normal(0, 0.5, len(idx))}, index=idx)
    df_vix = pd.DataFrame({'VIXCLS': 20 + np.random.normal(0, 5, len(idx))}, index=idx)

# World Bank (Dívida/PIB do Brasil)
# Handle potential API errors by assigning an empty DataFrame
try:
    df_wb_debt = wb.get_indicator('GC.DOD.TOTL.GD.ZS', 'BR', 2010, 2024)
except Exception as e: # Catching a broader exception to handle APIResponseError from wbgapi
    print(f"⚠️ Erro ao coletar dados do World Bank: {e}. Gerando DataFrame vazio para df_wb_debt.")
    df_wb_debt = pd.DataFrame()

print("✅ Dados coletados.")
```

## 🔗 Célula 6: Consolidação e Engenharia de Features

```python
import pandas as pd
import numpy as np # Ensure numpy is imported if not already

# Juntar todos os DataFrames em frequência mensal
dfs = [df_selic, df_ipca, df_cambio, df_fsi, df_fedfunds, df_vix]
df_macro = pd.concat(dfs, axis=1, join='outer')

# Renomear colunas BCB para nomes amigáveis
rename_map = {
    'bcb_4390': 'SELIC',
    'bcb_433': 'IPCA_mensal',
    'bcb_1': 'BRL_USD',
    'bcb_24375': 'FSI'
}
df_macro = df_macro.rename(columns=rename_map)

# Adicionar dados anuais do World Bank (preenchendo para meses)
# O df_wb_debt é garantido como DataFrame (potencialmente vazio) pela célula anterior.
if not df_wb_debt.empty:
    wb_series = df_wb_debt.iloc[:, 0].squeeze() # Use squeeze to ensure it's a Series
    wb_series = wb_series.resample('MS').ffill()
    df_macro['WB_DEBT_GDP'] = wb_series
else:
    print("⚠️ Dados do World Bank (Dívida/PIB) ausentes ou com erro. Gerando dados sintéticos para WB_DEBT_GDP.")
    # Gerar dados sintéticos para WB_DEBT_GDP com base no índice atual de df_macro
    # Assume-se que df_macro já possui um índice temporal válido a partir de outras fontes
    if not df_macro.empty:
        synthetic_wb_gdp = pd.Series(
            80 + np.linspace(-5, 5, len(df_macro.index)) + np.random.normal(0, 2, len(df_macro.index)),
            index=df_macro.index,
            name='WB_DEBT_GDP'
        )
        df_macro['WB_DEBT_GDP'] = synthetic_wb_gdp
    else:
        # Fallback if df_macro is unexpectedly empty at this point, though unlikely given BCB/FRED data
        print("⚠️ df_macro está inesperadamente vazio antes da geração de dados sintéticos do WB. WB_DEBT_GDP será preenchido com NaN.")
        df_macro['WB_DEBT_GDP'] = np.nan # If df_macro is somehow empty, cannot generate aligned synthetic data

# Calcular IPCA 12 meses e Juro Real
if 'IPCA_mensal' in df_macro.columns and 'SELIC' in df_macro.columns:
    df_macro['IPCA_12m'] = (1 + df_macro['IPCA_mensal'] / 100).rolling(12).apply(np.prod, raw=True) - 1
    df_macro['IPCA_12m'] *= 100
    df_macro['JURO_REAL'] = ((1 + df_macro['SELIC']/100) / (1 + df_macro['IPCA_12m']/100) - 1) * 100
else:
    print("⚠️ Colunas 'IPCA_mensal' ou 'SELIC' ausentes para cálculo de 'IPCA_12m' e 'JURO_REAL'.")
    # Inicializa com NaNs se as colunas não existirem para evitar mais erros
    if 'IPCA_12m' not in df_macro.columns: df_macro['IPCA_12m'] = np.nan
    if 'JURO_REAL' not in df_macro.columns: df_macro['JURO_REAL'] = np.nan

# Preencher valores ausentes e remover linhas com NaNs que impactam os cálculos
# dropna() deve ser feito APÓS a geração de dados sintéticos para WB_DEBT_GDP
df_macro = df_macro.ffill().dropna()

print(f"✅ DataFrame consolidado: {len(df_macro)} meses.")
df_macro.tail()
```

## 📈 Célula 7: Modelagem Econométrica (VECM e SPJ)

```python
import statsmodels.api as sm
from statsmodels.tsa.vector_ar.vecm import coint_johansen, VECM
from statsmodels.tsa.api import VAR
import statsmodels.formula.api as smf

# Inicializa variáveis para garantir que estejam definidas mesmo se o modelo não for executado
rank = None
model_type = None
spj_results = pd.DataFrame()

# Check if required columns exist before selecting
required_model_cols = ['WB_DEBT_GDP', 'JURO_REAL', 'IPCA_12m']
if all(col in df_macro.columns for col in required_model_cols):
    df_model = df_macro[required_model_cols].dropna()

    if not df_model.empty and len(df_model) > 10: # Ensure enough data points for modeling
        # Teste de Johansen
        try:
            joh = coint_johansen(df_model, det_order=0, k_ar_diff=1)
            rank = (joh.lr1 > joh.cvt[:, 1]).sum()
            print(f"🔗 Rank de cointegração: {rank}")

            # Estimar VECM ou VAR
            if rank > 0:
                model = VECM(df_model, k_ar_diff=2, coint_rank=rank, deterministic='ci')
                result = model.fit()
                model_type = 'VECM'
            else:
                model = VAR(df_model.diff().dropna()) # Ensure no NaNs from diff() before VAR
                result = model.fit(maxlags=2)
                model_type = 'VAR'

            print(f"✅ Modelo {model_type} estimado.")

            # ---------- SPJ (Split-Panel Jackknife) ----------
            def run_spj(df, formula):
                # Ensure all variables in the formula exist in df
                formula_vars = [var.strip() for var in formula.replace('~', '+').split('+') if var.strip()]
                if not all(v in df.columns for v in formula_vars):
                    print(f"⚠️ Variáveis da fórmula SPJ ausentes no DataFrame: {formula_vars}")
                    return pd.DataFrame() # Return empty DataFrame

                T = len(df)
                if T < 2: # Need at least 2 data points for split
                    print("⚠️ Dados insuficientes para SPJ.")
                    return pd.DataFrame()
                half = T // 2

                try:
                    full = smf.ols(formula, data=df).fit()
                    h1 = smf.ols(formula, data=df.iloc[:half]).fit()
                    h2 = smf.ols(formula, data=df.iloc[half:]).fit()
                    spj = 2 * full.params - 0.5 * (h1.params + h2.params)
                    return pd.DataFrame({'OLS': full.params, 'SPJ': spj, 'Gap': spj - full.params})
                except Exception as e:
                    print(f"⚠️ Erro ao rodar SPJ: {e}")
                    return pd.DataFrame()

            spj_results = run_spj(df_macro, 'IPCA_12m ~ WB_DEBT_GDP + JURO_REAL')
            print("\n📊 Resultados SPJ:")
            display(spj_results.round(4))
        except Exception as e:
            print(f"⚠️ Erro durante a estimação do modelo econométrico: {e}. Skipping VECM/VAR e SPJ.")
    else:
        print("⚠️ DataFrame para modelagem (df_model) está vazio ou tem poucos dados após dropna(). Skipping VECM/VAR e SPJ.")
else:
    print(f"⚠️ Colunas necessárias para modelagem {required_model_cols} ausentes no df_macro. Skipping VECM/VAR e SPJ.")
```

## 🔍 Célula 8: Análise do Fiscal Kink Point (Python Puro)

```python
from scipy.optimize import minimize_scalar
import matplotlib.pyplot as plt

def ssr_two_lines(kink, x, y):
    """Soma dos quadrados dos resíduos para duas regressões lineares."""
    mask = x <= kink
    if mask.sum() < 2 or (~mask).sum() < 2: # Need at least 2 points in each segment
        return np.inf
    # Ensure there are enough points for polyfit
    if len(x[mask]) < 2: return np.inf
    coef1 = np.polyfit(x[mask], y[mask], 1)
    y_pred1 = np.polyval(coef1, x[mask])
    ssr1 = np.sum((y[mask] - y_pred1)**2)

    if len(x[~mask]) < 2: return np.inf
    coef2 = np.polyfit(x[~mask], y[~mask], 1)
    y_pred2 = np.polyval(coef2, x[~mask])
    ssr2 = np.sum((y[~mask] - y_pred2)**2)
    return ssr1 + ssr2

# Inicializa variáveis para garantir que estejam definidas mesmo se o kink point não for detectado
kink_idx = None
kink_value = None
kink_date = None
fsi_series = pd.Series() # Define fsi_series as empty Series initially

# Usar a série FSI para detectar kink point
if 'FSI' in df_macro.columns and not df_macro['FSI'].empty:
    fsi_series = df_macro['FSI'].dropna() # Keep as Series for index
    if len(fsi_series) > 10: # Ensure enough data points for kink detection (e.g., min 5 for each segment)
        x = np.arange(len(fsi_series))
        y = fsi_series.values

        # Otimização
        try:
            # Ensure bounds are valid for the given data length
            min_bound = min(5, len(x) - 6) # at least 5 points for first segment, and 2 for second
            max_bound = max(5, len(x) - 6) # at least 2 points for first segment, and 5 for second

            if min_bound > max_bound:
                print("⚠️ Dados insuficientes para encontrar um intervalo válido para otimização do Kink Point.")
            else:
                result = minimize_scalar(ssr_two_lines, bounds=(min_bound, max_bound), args=(x, y), method='bounded')
                kink_idx = int(result.x)
                kink_value = fsi_series.iloc[kink_idx] # Access value by integer position
                kink_date = fsi_series.index[kink_idx]

                print(f"📍 Fiscal Kink Point identificado:")
                print(f"   - Índice: {kink_idx}")
                print(f"   - Valor do FSI: {kink_value:.4f}")
                print(f"   - Data aproximada: {kink_date.strftime('%Y-%m-%d')}")

                # Visualização
                plt.figure(figsize=(12,5))
                plt.plot(fsi_series.index, fsi_series, label='FSI') # Plot using datetime index
                plt.axvline(x=kink_date, color='red', linestyle='--', label=f'Kink Point ({kink_date.date()})')
                plt.title('Fiscal Stress Index e Ponto de Inflexão Fiscal')
                plt.legend()
                plt.grid(alpha=0.3)
                plt.show()
        except Exception as e:
            print(f"⚠️ Erro ao calcular Fiscal Kink Point: {e}")
            kink_idx = None
            kink_value = None
            kink_date = None
    else:
        print("⚠️ Série FSI tem dados insuficientes para detecção de Kink Point. Pelo menos 10 pontos são necessários.")
else:
    print("⚠️ Coluna 'FSI' ausente no df_macro ou está vazia. Skipping Fiscal Kink Point analysis.")
```
